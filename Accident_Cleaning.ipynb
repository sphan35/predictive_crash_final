```python
# Import dependencies 
import pandas as pd
import numpy as np
```

```python
# Read Csv for accident data
accident_data = pd.read_csv(r'/Users/dylankurth/Documents/Dev_Projects/predictive_crash_final/Data/1_FortBendCrashesAdjusted.csv')
accident_data.head()
```

```python
# Check rows
accident_data.count()
```

```python
# Drop columns that have too many null values 
accident_data.drop(['Curve_Lngth', 'Cd_Degr'], axis=1)
```

```python
# View datatype
accident_data['Curve_Type_ID'].dtype
```

```python
# Fill NaN
accident_data['Curve_Type_ID']=accident_data['Curve_Type_ID'].fillna(0)
```

```python
# Covert datatype
accident_data['Curve_Type_ID']=accident_data['Curve_Type_ID'].astype(int)
```

```python
# Conver values greater than 0 to 1
accident_data['Curve_Type_ID'].values[accident_data['Curve_Type_ID'].values > 0] = 1
```

```python
# Check new row count 
accident_data.count()
```
